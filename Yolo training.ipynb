# Pre-processing data


```python
import pandas as pd
import os
import shutil
import random


```

```python
thisPath="/Users/pCd/Downloads/Challenge/smartathon/Them1/dataset/"
train_dataset=pd.read_csv(thisPath+'dataset/train.csv')
test_dataset=pd.read_csv(thisPath+'dataset/test.csv')

```

**1. Seperate the test dataset into "output" folder**

```python
for img_path in test_dataset.values:
    print(img_path[0])
    
    shutil.move(thisPath+"dataset/images/"+img_path[0], thisPath+"dataset/images/output") 
```

**2. deviding them into train and validation set**

```python
for classN, image_path, name, xmax, xmin, ymax, ymin in train_dataset.values:
    try:
        if random.random() <0.20:
            print("val:"+image_path)
        
            shutil.move(thisPath+"dataset/images/"+image_path, thisPath+"dataset/images/val") 
        else:
            print("train:"+image_path)
            shutil.move(thisPath+"dataset/images/"+image_path, thisPath+"dataset/images/train") 
    except:
        print("Missing:"+image_path)
```

**3. generate the lableing files in YOLO7 format using the provided file**

```python
directory_path_train = thisPath+'dataset/images/train'
directory_files_train = os.listdir(directory_path_train)

directory_path_val = thisPath+'dataset/images/val'
directory_files_val = os.listdir(directory_path_val)
```

```python
imageHight=1080
imageWidth=1920
i=0
j=0
for classN, image_path, name, xmax, xmin, ymax, ymin in train_dataset.values:
    
    centerX= 2*((xmax+ xmin)/2)/imageWidth
    centerY= 2*((ymax+ ymin)/2)/imageHight
    width= 2*(xmax-xmin)/imageWidth
    height= 2*(ymax-ymin)/imageHight
    output= str(int(classN))+" "+str(centerX)+" "+str(centerY)+" "+str(width)+" "+str(height)
    
    if(image_path in directory_files_train):
       i+=1
       try:
          with open(thisPath+"/dataset/labels/train/"+image_path.split(".")[0]+".txt", "x") as file:
             file.write(output+"\n")
       except:
          with open(thisPath+"/dataset/labels/train/"+image_path.split(".")[0]+".txt", "a") as file:
              file.write(output+"\n")

    if (image_path in directory_files_val):
       j+=1
       try:
           with open(thisPath+"/dataset/labels/val/"+image_path.split(".")[0]+".txt", "x") as file:
              file.write(output+"\n")
       except:
           with open(thisPath+"/dataset/labels/val/"+image_path.split(".")[0]+".txt", "a") as file:
              file.write(output+"\n")
    
```

**2. convert the results into the submission format**

```python
# \runs\detect\exp\labels
directory_path_lables = thisPath+'yolov7/runs/detect/exp/labels'
directory_files_lables = os.listdir(directory_path_lables)
```

```python
clases=['GRAFFITI', 'FADED_SIGNAGE', 'POTHOLES', 'GARBAGE', 'CONSTRUCTION_ROAD', 'BROKEN_SIGNAGE', 'BAD_STREETLIGHT', 'BAD_BILLBOARD', 'SAND_ON_ROAD', 'CLUTTER_SIDEWALK', 'UNKEPT_FACADE']  # class names

with open(thisPath+"submission_960x540.csv",'w') as f:
    f.write('class,image_path,name,xmax,xmin,ymax,ymin\n')
    for file in directory_files_lables:
        with open(directory_path_lables+'/'+file,'r') as labels:
            lines = labels.readlines()
            for line in lines:
                line = line.split()
                
                classN=float(line[0])
                image_path= file.split('.')[0]+'.jpg'
                name=clases[int(classN)]
                xmax = int(1920* (float(line[1])+float(line[3])))
                xmin = int(1920* (float(line[1])-float(line[3])))
                ymax = int(1080* (float(line[2])+float(line[4])))
                ymin = int(1080* (float(line[2])-float(line[4])))
                f.write(str(classN) +','+ image_path+','+  name+','+  str(float(xmax))+','+  str(float(xmin))+','+  str(float(ymax))+','+  str(float(ymin))+'\n')
                print(str(classN) +','+ image_path+','+  name+','+  str(float(xmax))+','+  str(float(xmin))+','+  str(float(ymax))+','+  str(float(ymin)))
                
              

```
